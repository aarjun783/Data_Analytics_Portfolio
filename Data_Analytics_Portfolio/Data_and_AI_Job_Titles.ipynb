# Cleaning the list of 700 Data and AI roles

### Importing the dataset from kaggle and saving it as a pandas dataframe

```python
import mlcroissant as mlc
import pandas as pd

# Fetch the Croissant JSON-LD
croissant_dataset = mlc.Dataset('https://www.kaggle.com/datasets/princekhunt19/700-jobs-data-of-ai-and-data-fields-2025/croissant/download')

# Check what record sets are in the dataset
record_sets = croissant_dataset.metadata.record_sets
print(record_sets)

# Fetch the records and put them in a DataFrame
record_set_df = pd.DataFrame(croissant_dataset.records(record_set=record_sets[0].uuid))
record_set_df.head()
```

# Data Cleaning

### Removing special characters from dataframe columns

```python
#Remove special characters and prefixes from columns
record_set_df.columns = record_set_df.columns.str.replace('jobs_dataset.csv/','')
record_set_df.columns = record_set_df.columns.str.replace('%2F','_')
record_set_df.head()
```

### Removing metadata and url columns

```python
record_set_df = record_set_df.drop(columns=['url','externalApplyLink','searchInput_country','searchInput_position'])
record_set_df.head()
```

### Removing special characters and single quotes from records of object data type

```python
# Remove 'b'' and the trailing single quote from object columns
for col in record_set_df.select_dtypes(include='object').columns:
    record_set_df[col] = record_set_df[col].astype(str).str.replace("b'", "", regex=False).str.replace("'", "", regex=False)
# Remove empty space characters from each job description
record_set_df['description'] = record_set_df['description'].astype(str).str.replace("\n", " ", regex=False)
display(record_set_df.head())
```

#Data manipulation
Splitting location into city and state and removing postal codes if found. Identifying remote roles and saving them to the state column

```python
#Create state column from location column and keep cities in location column, keep remote value in location column
record_set_df['state'] = record_set_df['location'].str.split(', ').str[-1].replace(r'\d+', '', regex=True).str.strip()
record_set_df['location'] = record_set_df['location'].str.split(', ').str[:-1].str.join(', ')
record_set_df.loc[record_set_df['state'].str.contains('remote', case=False, na=False), 'location'] = 'Remote'
record_set_df.loc[record_set_df['location'] == 'remote', 'state'] = None
display(record_set_df.head())
```

Classifying the rest of the unabbreviated columns into the state abbreviations. For future versions that contain unabbreviated data, the dictionary will have to be updated

```python
state_abbreviations = {
    'California': 'CA',
    'Florida': 'FL',
    'Maryland': 'MD',
    'Minnesota': 'MN',
    'New Jersey': 'NJ',
    'New York State': 'NY',
    'North Carolina': 'NC',
    'Virginia': 'VA',
    'Washington State': 'WA',
    'Wisconsin': 'WI'
}

record_set_df['state'] = record_set_df['state'].replace(state_abbreviations)
```

### Extracting the average annual salary from the salary ranges in the salary column.
If a record contains hourly salary ranges or salaries, it must be converted into annual salaries

```python
import re
#Create function for extracting average yearly salary
def process_salary(salary):
    if isinstance(salary, str):
        salary = salary.lower()
        # Handle hourly rates and convert to yearly
        if 'hour' in salary:
            hourly_rate_match = re.search(r'(\d+(\.\d+)?)\s*(-\s*(\d+(\.\d+)?))?', salary)
            if hourly_rate_match:
                rate1 = float(hourly_rate_match.group(1)) #Grab minimum value
                rate2 = float(hourly_rate_match.group(4)) if hourly_rate_match.group(4) else rate1 #Grab maximum value else minimum value
                avg_hourly_rate = (rate1 + rate2) / 2 #Calculate the average hourly salary
                # Assuming a standard work year of 40 hours/week * 52 weeks/year
                return avg_hourly_rate * 40 * 52
        # Handle yearly salary ranges and calculate average
        elif 'year' in salary or ('$' in salary and '-' in salary):
            salary_range_match = re.search(r'\$(\d{1,3}(,\d{3})*(\.\d+)?)\s*-\s*\$(\d{1,3}(,\d{3})*(\.\d+)?)', salary)
            if salary_range_match:
                salary1 = float(salary_range_match.group(1).replace(',', ''))
                salary2 = float(salary_range_match.group(4).replace(',', ''))
                return (salary1 + salary2) / 2
            # Handle single yearly salary values
            single_salary_match = re.search(r'\$(\d{1,3}(,\d{3})*(\.\d+)?)', salary)
            if single_salary_match:
                return float(single_salary_match.group(1).replace(',', ''))

    return None # Return None for values that don't match expected formats

record_set_df['estimated_yearly_salary'] = record_set_df['salary'].apply(process_salary)
record_set_df['estimated_yearly_salary'] = record_set_df['estimated_yearly_salary'].fillna(0)
```

### Extracting job titles from the position column.
Fuzzy string matching libraries like rapidfuzz provides approximate posititon role match with specific job titles rather than absolute role matches.
Regex was also used to split a role into multiple roles and remove seniority levels, departments and abbreviations of products

```python
#Extract and classify titles
import rapidfuzz as rf
from rapidfuzz import process
import re

canonical_titles_dict = {
    'Data Scientist': ['Data Science', 'Data Scientist', 'Data Scientist, Research'],
    'Data Analyst': ['Data Analyst', 'Analytics Analyst','Data Visualization'],
    'Data Engineer': ['Data Engineer', 'Analytics Engineer'],
    'AI Engineer': ['AI Engineer', 'Artificial Intelligence Engineer', 'Machine Learning Engineer','AI Developer','AI Researcher'],
    'ML Engineer': ['ML Engineer', 'Machine Learning Engineer'],
    'Applied Scientist': ['Applied Scientist', 'Research Scientist'],
    'Analytics Lead': ['Analytics Lead','Advanced Analytics Lead'],
    'Software Engineer': ['Software Engineer'],
    'Research Scientist': ['Research Scientist', 'Applied Scientist'],
    'NLP Engineer': ['NLP Engineer', 'Natural Language Processing Engineer'],
    'Data Architect': ['Data Architect'],
    'Head of Data and AI': ['Head of Data and ML','Head of AI'],
    'Product Analyst': [
        'product analyst', 'growth analyst', 'conversion analyst'
    ],
    'Tech Strategist':['Tech Strategist','AI Strategist'],
    'Statistician':['Statistician','Applied Statistician'],
    'Product Manager':['Product Manager','Product Lead']
}

def fuzzy_classify_title(raw_title, threshold=55):
    # process.extractOne returns None if no match is found above the score_cutoff
    all_aliases = {alias: canon for canon, aliases in canonical_titles_dict.items() for alias in aliases}
    match_result = process.extractOne(raw_title, all_aliases.keys(), score_cutoff=threshold)
    if match_result is None:
        return "other"
    else:
        match, score, _ = match_result # Corrected: unpack match_result directly
        return all_aliases[match] # Corrected: return the canonical title using the matched alias

def split_position(positionName):
    split = re.split(r'\s*/\s*|\s*&\s*|\s*and\s*|\s*[-|,]\s*', positionName.strip(), flags=re.IGNORECASE) #Split title into a list of strings
    split = [word.lower() for word in split]
    # Remove seniority levels, departments, abbreviations, and specific product names
    cleaned = [re.sub(r'(senior|lead|principal|jr\.?|sr\.?|mid-level|entry-level|youtube|aws)', '', word) for word in split]
    cleaned = [word.strip() for word in cleaned]
    # Apply fuzzy classification to each cleaned word
    classified_titles = [fuzzy_classify_title(word) for word in cleaned]
    # Remove "other" and duplicates
    filtered_titles = []
    for title in classified_titles:
        if title != "other" and title not in filtered_titles:
            filtered_titles.append(title)
    return filtered_titles


record_set_df['positionName'] = record_set_df['positionName'].apply(split_position)
display(record_set_df.head())
```

### Creating two new skills lists for extracting technical and soft skills

```python
technical_skills = [
    # Programming & Data Wrangling
    'Python', 'SQL', 'R', 'Excel', 'NumPy', 'Pandas','C','C++',

    # Machine Learning & AI
    'Machine Learning', 'Deep Learning', 'Reinforcement Learning',
    'TensorFlow', 'PyTorch', 'Scikit-learn', 'XGBoost',
    'Natural Language Processing', 'NLP', 'Computer Vision',
    'Generative AI', 'Agentic AI',

    # Cloud & Deployment
    'AWS', 'Azure', 'GCP', 'Docker', 'Kubernetes', 'Streamlit', 'Flask',

    # Big Data & Distributed Systems
    'Spark', 'Hadoop', 'Kafka', 'Snowflake',

    # Data Engineering
    'ETL', 'Data Warehousing', 'dbt', 'Airflow', 'BigQuery', 'Redshift',

    # Business Intelligence
    'Power BI', 'Tableau', 'Looker', 'Business Intelligence',

    # Other
    'Statistical Modeling', 'Data Mining', 'Big Data'
]

soft_skills = [
    'Communication', 'Teamwork', 'Collaboration', 'Problem Solving', 'Critical Thinking',
    'Adaptability', 'Leadership', 'Interpersonal Skills', 'Presentation', 'Negotiation',
    'Time Management', 'Mentoring', 'Curiosity', 'Stakeholder Management',
    'Attention to Detail', 'Self-motivation', 'Accountability', 'Creativity'
]
```

### Extracting hard and soft skills from the description, and creating two new columns in the dataframe

```python
def extract_skills_from_description(description):
    found_technical_skills = []
    found_soft_skills = []

    if isinstance(description, str):
        description_lower = description.lower()
        for skill in technical_skills:
            if skill.lower() in description_lower:
                found_technical_skills.append(skill)
        for skill in soft_skills:
            if skill.lower() in description_lower:
                found_soft_skills.append(skill)

    return found_technical_skills, found_soft_skills

record_set_df['identified_technical_skills'], record_set_df['identified_soft_skills'] = zip(*record_set_df['description'].apply(extract_skills_from_description))
display(record_set_df.head())
```

### Flattening the dataset for EDA and dashboards

```python
record_set_df['Title_Key'] = record_set_df.index+1
record_set_df = record_set_df.explode('positionName')
record_set_df = record_set_df.explode('identified_technical_skills')
record_set_df = record_set_df.explode('identified_soft_skills')

# Get the list of current columns
current_columns = record_set_df.columns.tolist()

# Remove 'Title_Key' from the list
current_columns.remove('Title_Key')

# Create a new list with 'Title_Key' as the first column
new_column_order = ['Title_Key'] + current_columns

# Reindex the DataFrame with the new column order
record_set_df = record_set_df[new_column_order]


display(record_set_df.head())
```
